```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from collections import defaultdict
from statsmodels.formula.api import ols, mixedlm
import numpy as np
import statsmodels.api as sm
import pandas as pd
from itertools import combinations
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import spearmanr
from matplotlib.patches import Rectangle
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
from scipy.stats import pearsonr, linregress
from matplotlib.colors import LinearSegmentedColormap


from matplotlib.transforms import Affine2D

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
from statsmodels.formula.api import ols, mixedlm
from sklearn.preprocessing import MaxAbsScaler, RobustScaler, StandardScaler
from scipy.stats import pointbiserialr, spearmanr

from copy import copy
from tqdm import *
from glob import glob


scaler = StandardScaler()


def rainbow_text(
    x, y, strings, colors, bolds=None, ax=None, fontsizes=None, transform=None, **kwargs
):
    """
    Take a list of *strings* and *colors* and place them next to each
    other, with text strings[i] being shown in colors[i].

    Parameters
    ----------
    x, y : float
        Text position in data coordinates.
    strings : list of str
        The strings to draw.
    colors : list of color
        The colors to use.
    ax : Axes, optional
        The Axes to draw into. If None, the current axes will be used.
    **kwargs
        All other keyword arguments are passed to plt.text(), so you can
        set the font size, family, etc.
    """
    bolds = bolds or []
    if ax is None:
        ax = plt.gca()
    t = ax.transAxes if transform is None else transform
    canvas = ax.figure.canvas

    for idx, (s, c) in enumerate(zip(strings, colors)):
        fontweight = "bold" if idx in bolds else "normal"
        fontsize = None
        if fontsizes is None and "fontsize" in kwargs:
            fontsize = kwargs.pop("fontsize")
        elif fontsizes is not None:
            fontsize = fontsizes[idx]

        text = ax.text(
            x,
            y,
            s + " ",
            color=c,
            transform=t,
            ha="left",
            fontweight=fontweight,
            fontsize=fontsize,
            **kwargs
        )

        # Need to draw to update the text position.
        text.draw(canvas.get_renderer())
        ex = text.get_window_extent()
        t = text.get_transform() + Affine2D().translate(ex.width, 0)


palette = sns.color_palette("Paired", n_colors=20)
plt.rcParams["axes.prop_cycle"] = plt.cycler("color", palette)

plt.rcParams["lines.solid_capstyle"] = "projecting"

sns.set_theme(style="whitegrid", font_scale=1.2, palette="tab10")
plt.rcParams["lines.solid_capstyle"] = "projecting"
matplotlib.rcParams["axes.unicode_minus"] = False
matplotlib.rc("hatch", color="1.0", linewidth=0.5)
matplotlib.rc("font", **{"family": "sans-serif", "sans-serif": ["Helvetica"]})

plt.rcParams["font.family"] = "Helvetica"


model_name_dict = {
    "gpt2large": "GPT2-large",
    "zephyr": "Zephyr",
    "llama-2": "Llama-2",
    "llama-2-chat": "Llama-2-chat",
    "falcon-7b": "Falcon-7b",
    "falcon-7b-instruct": "Falcon-7b-instruct",
    "mistral7b": "Mistral-7b",
    "mistral7b-instruct": "Mistral-7b-instruct",
}

pprint_condition = {
    "LEXBOOST_nRand": "Random Noun Overlap",
    "LEXBOOST_allN": "All Nouns Overlap",
    "LEXBOOST_Vonly": "Verb + Prep. Overlap",
    "LEXBOOST_allfunc": "Det. + Prep. Overlap",
    "LEXBOOST_det": "Determiners Overlap",
    "LEXBOOST_prep": "Preposition Overlap",
    "SEMSIM_Nall": "Nouns More Similar",
    "SEMSIM_Vonly": "Verbs More Similar",
    "SEMSIM_all": "Nouns+Verbs More Similar",
    "ANOMALOUS": "Anomalous",
    "CORE": "Core",
}

pprint_vars = {
    "n1_sims": "sim($n_1$)",  # ^p$, $n_1^t$)',
    "n2_sims": "sim($n_2$)",  # ^p$, $n_2^t$)',
    "n3_sims": "sim($n_3$)",  # ^p$, $n_3^t$)',
    "verb_sims": "sim($v$)",  # $, $v^t$)',
    "sen_sim_x": "sim($s$)",  # $, $s^t$)',
    "logp_prime_x": "$-$P(prime$_{po}$)",
    "logp_prime_y": "$-$P(prime$_{do}$)",
    "logp_x": "$-$P(target$_{po}$)",
    "logp_y": "$-$P(target$_{do}$)",
    "average_verb_preference_prime": "PO-pref($v^p$)",
    "average_verb_preference_target": "PO-pref($v^t$)",
    "n1_n1_overlap": "$N_1$ overlaps",
    "n2_n2_overlap": "$N_2$ overlaps",
    "n3_n3_overlap": "$N_3$ overlaps",
    "det_overlap": "Det. overlaps",
    "pp_overlap": "Prep. overlaps",
    "verb_overlap": "Verb overlaps",
}
```

# Semantic Similarity

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from nltk.stem import WordNetLemmatizer
import nltk
import numpy as np
from numpy.linalg import norm
import torch

lemmatizer = WordNetLemmatizer()

model = AutoModelForCausalLM.from_pretrained('gpt2-large')
model_embs = model.transformer.wte.weight

# model_embs = torch.load('../../scripts/gpt2-large_embs.pt')
tokenizer = AutoTokenizer.from_pretrained('gpt2-large')


def lemma(word, tag):
    return lemmatizer.lemmatize(word, tag)


class Embs:
    def __init__(self, embs):
        self.embs = embs 

    def __getitem__(self, key):
        w, t = key

        word = 'Ġ' + w

        token_idx = tokenizer.convert_tokens_to_ids(word)
        
        if token_idx == tokenizer.unk_token_id:
            word = 'Ġ' + lemmatizer.lemmatize(w, t)
            token_idx = tokenizer.convert_tokens_to_ids(word)
        if token_idx == tokenizer.unk_token_id:
            word = 'Ġ' + lemmatizer.lemmatize(w, t)
            token_idx = tokenizer.convert_tokens_to_ids(word)
        
        assert token_idx != tokenizer.unk_token_id, w
        
        return self.embs[token_idx].detach().numpy()

embs = Embs(model_embs)

def cosine(w1, w2, tag):
    return (embs[w1, tag] @ embs[w2, tag]) / (norm(embs[w1, tag]) * norm(embs[w2, tag]))
```

```python
bigram_sims = {}
token_sim_dict = {
    "verb": np.zeros(len(mega_df)),
    "n1": np.zeros(len(mega_df)),
    "n2": np.zeros(len(mega_df)),
    "n3": np.zeros(len(mega_df)),
}

token_tags = [
    ('verb', 'v', verb_sims),
    ('n1', 'n', n1_sims),
    ('n2', 'n', n2_sims),
    ('n3', 'n', n3_sims),
]

for idx in trange(len(mega_df)):
    row = mega_df.iloc[idx]
    
    for token, tag, sims in token_tags: 
        prime_token, target_token = sorted((row[f'prime_{token}'], row[f'target_{token}']))

        if (prime_token, target_token) in bigram_sims:
            token_sim_dict[token][idx] = token_sim_dict[prime_token, target_token]
        else:
            sim = cosine(prime_token, target_token, tag)
            bigram_sims[prime_token, target_token] = sim
            token_sim_dict[token][idx] = sim

mega_df['verb_sims'] = verb_sims
mega_df['n1_sims'] = n1_sims
mega_df['n2_sims'] = n2_sims
mega_df['n3_sims'] = n3_sims
```

### Sentence Similarity

```python
from sentence_transformers import SentenceTransformer
from numpy import dot
from numpy.linalg import norm


def cos_sim_np(a, b):
    return dot(a, b)/(norm(a)*norm(b))


sentence_encoder = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
from tqdm import *
import numpy as np


sen2sim = {}
sensims = []

df = mega_df[mega_df.model_name == 'gpt2large']

all_prime = [
    prime.replace('<|endoftext|>', '').strip() 
    for prime in df.prime_x
]
all_target = [
    target.replace('<|endoftext|>', '').strip() 
    for target in df.x
]

prime_embeddings = []
target_embeddings = []

batch_size = 64
num_batches = len(df) / (batch_size-1)

for prime_batch in tqdm_notebook(np.array_split(all_prime, num_batches)):
    batch_embeddings = sentence_encoder.encode(prime_batch)
    prime_embeddings.extend(batch_embeddings)

for target_batch in tqdm_notebook(np.array_split(all_target, num_batches)):
    batch_embeddings = sentence_encoder.encode(target_batch)
    target_embeddings.extend(batch_embeddings)

for prime_target_embs in tqdm_notebook(zip(prime_embeddings, target_embeddings)):
    sensims.append(cos_sim_np(*prime_target_embs))
```

```python
for prime, target, sensim in zip(all_prime, all_target, sensims):
    sen2sim[prime, target] = sensim
    
all_sensims = []
mean_sim = np.mean(sensims)

for prime, target, model_name, condition in zip(mega_df.prime_x, mega_df.x, mega_df.model_name, mega_df.condition):
    prime = prime.replace('<s>','').replace('<|endoftext|>','').strip()
    target = target.replace('<s>','').replace('<|endoftext|>','').strip()
    sensim = sen2sim.get((prime, target), mean_sim)
    
    all_sensims.append(sensim)

mega_df['sen_sim_x'] = np.array(all_sensims)
```

```python
model2color = {
    'falcon-7b': sns.color_palette('Paired', n_colors=8)[1], 
    'falcon-7b-instruct': sns.color_palette('Paired', n_colors=8)[0],     
    'llama-2':  sns.color_palette('Paired', n_colors=8)[3],
    'llama-2-chat':  sns.color_palette('Paired', n_colors=8)[2],
    'gpt2large':  sns.color_palette('Paired', n_colors=12)[11],
    'mistral7b':  sns.color_palette('Paired', n_colors=8)[5],
    'mistral7b-instruct':  sns.color_palette('Paired', n_colors=8)[4],
    'zephyr':  sns.color_palette('Paired', n_colors=8)[6],
}

def plot_pe_scatter(
    ax_idx,
    ax, 
    conditions, 
    xlim, 
    ylim, 
    title=None, 
    plot_ylabel=False, 
    skip_models=None,
    pe_x_attr: str = 'pe_x',
    pe_y_attr: str = 'pe_y',
    show_gradient=False,
    plot_legend=False,
):
    if skip_models is None:
        skip_models = []

    xmin, xmax = xlim
    ymin, ymax = ylim

    # plt.plot([-1,1],[2,-2])
    ax.add_patch(Rectangle(xy=(xmin,ymin) ,width=-xmin, height=-ymin, facecolor='#ffe5e5', fill=True, alpha=1., zorder=-100))
    ax.add_patch(Rectangle(xy=(0,ymin) ,width=xmax, height=-ymin, facecolor='#ffffe5', fill=True, alpha=1., zorder=-100))
    ax.add_patch(Rectangle(xy=(0,0) ,width=xmax, height=ymax, facecolor='#e5f2e5', fill=True, alpha=1., zorder=-100))
    ax.add_patch(Rectangle(xy=(xmin,0) ,width=-xmin, height=ymax, facecolor='#ffffe5', fill=True, alpha=1., zorder=-100))   

    ax.plot([xmin, xmax], [ymin, ymax], ':', color='0.8', zorder=-2)

    for condition in conditions:
        all_core_pe_y = [
            (model_name, mega_df[(mega_df.condition == 'CORE') & (mega_df.model_name == model_name)][pe_y_attr].mean())
            for model_name in mega_df.model_name.unique()
            if model_name not in skip_models
        ]
        all_core_pe_y = sorted(all_core_pe_y, key=lambda x: x[1])
        min_pe = all_core_pe_y[0][1] / 2.5
        max_pe = all_core_pe_y[-1][1] + min_pe * 2.5
        model_order = [model_name for model_name, _ in all_core_pe_y]

        for model_name in mega_df.model_name.unique():
            if model_name in skip_models:
                continue

            df = mega_df[(mega_df.condition == condition) & (mega_df.model_name == model_name)]
            pe_x_mean = df[pe_x_attr].mean()
            pe_y_mean = df[pe_y_attr].mean()
            
            core_df = mega_df[(mega_df.condition == 'CORE') & (mega_df.model_name == model_name)]
            core_pe_x_mean = core_df[pe_x_attr].mean()
            core_pe_y_mean = core_df[pe_y_attr].mean()

            core_slope = linregress(core_df[pe_x_attr], core_df[pe_y_attr])[0]
            core_marker_size = spearmanr(core_df[pe_x_attr], core_df[pe_y_attr])[0]
            rho_size = .1
            core_rho_size = rho_size * (core_df[pe_x_attr].std() + core_df[pe_y_attr].std())
            ax.plot(
                [core_pe_x_mean-core_rho_size, core_pe_x_mean+core_rho_size], 
                [core_pe_y_mean-(core_slope*core_rho_size), core_pe_y_mean+(core_slope*core_rho_size)], 
                color="black",
                alpha=0.3,
                zorder=-1,
            )

            condition_slope = linregress(df[pe_x_attr], df[pe_y_attr])[0]
            condition_marker_size = spearmanr(df[pe_x_attr], df[pe_y_attr])[0]
            condition_rho_size = rho_size * (df[pe_x_attr].std() + df[pe_y_attr].std())
            ax.plot(
                [pe_x_mean-condition_rho_size, pe_x_mean+condition_rho_size], 
                [pe_y_mean-(condition_slope*condition_rho_size), pe_y_mean+(condition_slope*condition_rho_size)], 
                color="black",
                alpha=0.3,
                zorder=-1,
                clip_on=False,
            )

            color = model2color[model_name]
            scat = ax.scatter(
                pe_x_mean, 
                pe_y_mean, 
                label=model_name, 
                marker="p", 
                s=50,  #140*(condition_marker_size+1), 
                color=color,
            )
            
            ax.scatter(
                core_pe_x_mean, 
                core_pe_y_mean, 
                label=model_name, 
                marker="o", 
                s=50,  #140*(core_marker_size+1), 
                color=color,
            )
            
            model_index = model_order.index(model_name)
            label_y = model_index * ((max_pe-min_pe)/len(all_core_pe_y)) + min_pe
            
            ax.plot([xmin-.1, core_pe_x_mean], [label_y, core_pe_y_mean], ':', lw=1, alpha=1.0, clip_on=False, color=color)        
            ax.text(
                xmin-.1, 
                label_y, 
                model_name, 
                verticalalignment='center', 
                color=color, 
                fontsize=14, 
                horizontalalignment='right', 
                fontweight="bold",
            )
            ax.plot([core_pe_x_mean, pe_x_mean], [core_pe_y_mean, pe_y_mean], lw=1, alpha=.4, color=color)

    text_color = "0.7"
    y_offset = 0.1
    x_offset = 0.1
    fontsize = 11

    ax.text(xmin+x_offset, ymin+y_offset, 'inverse\npriming', color=text_color, fontsize=fontsize)
    ax.text(xmin+x_offset, 0+y_offset, 'skewed\npriming', color=text_color, fontsize=fontsize)
    if not plot_ylabel:
        ax.text(xmax-x_offset, ymin+y_offset, 'skewed\npriming', color=text_color, fontsize=fontsize, horizontalalignment='right')
        ax.text(xmax-x_offset, 0+y_offset, 'balanced\npriming', color=text_color, fontsize=fontsize, horizontalalignment='right')

    ax.set_ylim(ymin, ymax)
    ax.set_xlim(xmin, xmax)

    ax.set_xticks(list(np.arange(np.ceil(xmin), np.floor(xmax)+1)))

    tickcolor = '0.7'

    if plot_ylabel:
        ax.yaxis.tick_right()
        ax.yaxis.set_label_position("right")
        ax.set_yticks(list(np.arange(np.ceil(xmin), np.floor(xmax)+1)))
        ax.set_ylabel("$s$-PE(DO)")
    else:
        ax.set_yticks([])

    if plot_legend:
        legend_elements = [
            Line2D([0], [0], marker='p', color='0.7', label=0, markerfacecolor='0.3', markersize=14),
            Line2D([0], [0], marker='p', color='0.7', label=-0.5, markerfacecolor='0.3', markersize=10),
            Line2D([0], [0], marker='p', color='0.7', label=-1, markerfacecolor='0.3', markersize=6),
        ]

        legend = ax.legend(handles=legend_elements, loc='lower right', title=r'Spearman $\rho$')
        legend.get_frame().set_linewidth(0.0),
        plt.setp(legend.get_title(),fontsize=13)
    
    ax.spines['bottom'].set_color(tickcolor)
    ax.tick_params(axis='x', colors=tickcolor, length=0)
    ax.tick_params(axis='y', colors=tickcolor, length=0)

    ax.grid(False)
    
    ax.tick_params(
        axis='both', 
        which='major', 
        labelsize=12,  # 9 
        labelcolor="0.7",
        # pad=-3,
    )
    ax.set_xlabel('$s$-PE(PO)', fontsize=16)

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    rainbow_text(
        0., 
        1.15, 
        ["CDE"[ax_idx]+"."]+[' '.join(title[:2])], 
        ['0.5', '0.0'], 
        bolds=[1], 
        ax=ax, 
        fontsizes=[16,16],
    )
    rainbow_text(
        0., 
        1.05, 
        ["CDE"[ax_idx]+"."]+[' '.join(title[2:])], 
        ['1.0', '0.0'], 
        bolds=[1], 
        ax=ax, 
        fontsizes=[16,16],
    )

    
    # GRADIENTS
    if show_gradient:
        y2g = LinearSegmentedColormap.from_list("Custom", ['#ffffe5', 'white', '#e5f2e5'], N=100)
        r2y = LinearSegmentedColormap.from_list("Custom", ['#ffe5e5', 'white', '#ffffe5'], N=100)
        ax.imshow(
            [[0,10,20], [0,10,20], [0,10,20]],
            interpolation = 'bicubic',
            cmap=y2g,
            extent=[-.5,.5,0,ymax],
            zorder=-10,
        )
        ax.imshow(
            [[0,10,20], [0,10,20], [0,10,20]],
            interpolation = 'bicubic',
            cmap=r2y,
            extent=[-.5,.5,ymin,0],
            zorder=-10,
        )
        ax.imshow(
            [[20,20,20], [10,10,10], [0,0,2]],
            interpolation = 'bicubic',
            cmap=y2g,
            extent=[0,xmax,-.5,.5],
            zorder=-10,
        )
        ax.imshow(
            [[20,20,20], [10,10,10], [0,0,2]],
            interpolation = 'bicubic',
            cmap=r2y,
            extent=[xmin,0,-.5,.5],
            zorder=-10,
        )

        w2g = LinearSegmentedColormap.from_list("Custom", ['white', '#e5f2e5'], N=100)
        w2y = LinearSegmentedColormap.from_list("Custom", ['white', '#ffffe5'], N=100)
        w2r = LinearSegmentedColormap.from_list("Custom", ['white', '#ffe5e5'], N=100)

        ax.imshow(
            [[0,10,10], [0,10,10], [0,0,0]],
            interpolation = 'bicubic',
            cmap=w2g,
            extent=[0, .5, 0, .5],
            zorder=-10,
        )
        ax.imshow(
            [[10,0], [0,0]],
            interpolation = 'bicubic',
            cmap=w2y,
            extent=[-.5, 0, 0, .5],
            zorder=-10,
        )
        ax.imshow(
            [[0,0], [10,0]],
            interpolation = 'bicubic',
            cmap=w2r,
            extent=[-.5, 0, -.5, 0],
            zorder=-10,
        )
        ax.imshow(
            [[0,0], [0,10]],
            interpolation = 'bicubic',
            cmap=w2y,
            extent=[0, .5, -.5, 0],
            zorder=-10,
        )


conditions = ['SEMSIM_Vonly', 'SEMSIM_Nall', 'SEMSIM_all']
conditions = ['LEXBOOST_allN', 'LEXBOOST_prep', 'LEXBOOST_Vonly']
num_conditions = len(conditions)
fig, axes = plt.subplots(1, num_conditions, figsize=(num_conditions*5,4), facecolor='w')
ax_idx = 0

for condition in conditions:
    plot_pe_scatter(
        ax_idx,
        axes[ax_idx],
        [condition], 
        (-1.5, 4.1),  # (-1, 3.8), #
        (-1.5, 4.1),  # (-1, 3.8), #
        title=["Core", "($●$) $\\rightarrow$", f"{pprint_condition[condition]}" ,"($⬟$)"],
        plot_ylabel=(ax_idx==(num_conditions-1)),
        skip_models=['opt1.3b', 'pythia1.4b'],
        pe_x_attr='pe_x_from4',
        pe_y_attr='pe_y_from4',
    )
    ax_idx += 1

plt.subplots_adjust(wspace=.65)
plt.savefig('figures/lexboost_s-pe_nogradient.pdf', bbox_inches='tight')
plt.show()
```

```python
from typing import *
import numpy as np


def create_token_pe_plot(
    ax, 
    df, 
    x_attr: str, 
    y_attr: str, 
    pattern: List[str], 
    title: str, 
    num_pad: int = 0, 
    label = "",
    ylim = None,
    split_primed: bool = False,
    plot_ylabel: bool = True,
    skip_first_n: int = 0,
    tight_title: bool = False,
    colors=None,
    yticks=None,
    plot_delta_region=False,
    example_sen=None,
):
    subset_df = df[[len(x) == len(pattern) for x in df[x_attr]]]
    print(len(subset_df), len(df))
    token_pe = np.array(list(subset_df[x_attr])) - np.array(list(subset_df[y_attr]))

    tokens = []
    flat_pe = []
    primed = []

    pct_primed = 0.
    
    for row in token_pe:
        pct_primed += (sum(row) > 0) / len(token_pe)
        
        for idx, (token, pe) in enumerate(zip(pattern, row)):
            if idx < skip_first_n:
                continue

            tokens.append(token)
            flat_pe.append(pe)
            primed.append(sum(row) > 0)

        for pad_idx in range(num_pad):
            tokens.append(" "*pad_idx)
            flat_pe.append(0.0)
            primed.append(sum(row) > 0)

    avg_df = pd.DataFrame(data={"token": tokens, "token_pe": flat_pe, "primed": primed})

    # neg_pe_df = avg_df[avg_df.sentence_PEs <= 0]
    # pos_pe_df = avg_df[avg_df.sentence_PEs > 0]

    def colors_from_values(values, palette_name, max_value=None):
        # normalize the values to range [0, 1]
        # values *= -1
        if max_value is None and ylim is not None:
            max_value = max(map(abs, ylim))
        else:
            max_value = max(abs(values))
        normalized = ((values / max_value) + 1) / 2
        
        # normalized = (values - min(values)) / (max(values) - min(values))
        # convert to indices
        # use the indices to get the colors
        palette = sns.color_palette(palette_name, as_cmap=True)
        return palette(normalized)

    avg_token_pe = token_pe.mean(0)

    if split_primed:
        sns.barplot(ax=ax, data=avg_df, x='token', y='token_pe', hue='primed') 
    else:
        sns.barplot(ax=ax, data=avg_df, x='token', y='token_pe') 

    ax.axhline(0, c='0.4', linewidth=1.0)

    # GRID
    ax.grid(color='0.7', axis='y', linestyle='--', linewidth=0.5)

    # SPINES
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    plt.setp(ax.spines.values(), color="0.7", lw=.5)

    # TICKS & LABELS
    ax.tick_params(
        axis='x', 
        which='major', 
        labelsize=12,  # 9
        pad=-3,
        labelcolor="black",
    )
    ax.tick_params(
        axis='y', 
        which='major', 
        labelsize=12,  # 9 
        pad=-3,
        labelcolor="0.7",
    )

    ax.set_xlabel("")
    if plot_ylabel:
        pad = 0 if tight_title else -3
        ax.set_ylabel("$w$-PE", labelpad=pad)
    else:
        ax.set_yticklabels([])
        ax.set_ylabel("")

    if yticks is not None:
        ax.set_yticks(yticks)

    ax.set_xticks(range(len(pattern)+num_pad-skip_first_n), pattern[skip_first_n:] + [""]*num_pad)

    # TITLE
    avg_pe = -avg_token_pe.sum()
    if tight_title:
        title = title.replace("%pe", f"{-avg_pe:.2f}")
        rainbow_text(.05, 1.05, [title.split()[0], ' '.join(title.split()[1:])], ['black', '0.5'], ax=ax, fontsize=13) #10       
    else:
        title = title.replace("%pe", f"{-avg_pe:.2f}")
        rainbow_text(
            0., 
            1.05, 
            [label, title.split()[0], ' '.join(title.split()[1:])], 
            ['0.5', 'black', '0.5'], 
            ax=ax,
            bolds=[1],
            fontsizes=[14,15,13],
        )
    
    # YLIM
    if ylim is not None:
        ax.set_ylim(ylim)
        
    max_value = max([abs(bar.get_height()) for bar in ax.patches])/2
        
    for idx, bar in enumerate(ax.patches):
        if colors is None:
            bar.set_color(colors_from_values(bar.get_height(), 'coolwarm'))
            bar.set_edgecolor("1.0")
        elif isinstance(colors[idx], tuple):
            main_color, hatch_color = colors[idx]
            
            plt.rcParams["hatch.linewidth"]= 6

            bar.set_facecolor(main_color)
            bar.set_edgecolor(hatch_color)
            bar.set_hatch('//')
        else:
            bar.set_color(colors[idx])
            bar.set_edgecolor("1.0")
            
        if example_sen is not None:
            if idx < len(example_sen.split()):
                token = example_sen.split()[idx]
                ax.text(idx, -0.9, token, fontsize=9, style='italic', transform=ax.transData, ha='center', color='0.5')
        
        
    if split_primed:
        ax.get_legend().remove()
        
    if plot_delta_region:
        ymin, ymax = ylim
        # ax.add_patch(Rectangle(xy=(0,ymin) ,width=4, height=ymax-ymin, facecolor='red', fill=True, alpha=0.1, zorder=-100))
        ax.add_patch(Rectangle(xy=(3.5,ymin) ,width=5, height=ymax-ymin, facecolor='green', fill=True, alpha=0.1, zorder=-100))
        
    return avg_df, subset_df


def get_ylim(df, xlen, ylen):
    subset_df = df[[len(x) == xlen for x in df.all_logp_x_px]]
    px_token_pe = np.array(list(subset_df.all_logp_x_px)) - np.array(list(subset_df.all_logp_x_py))
    
    subset_df = df[[len(x) == ylen for x in df.all_logp_y_py]]
    py_token_pe = np.array(list(subset_df.all_logp_y_py)) - np.array(list(subset_df.all_logp_y_px))

    ymax = max(max(abs(px_token_pe.mean(0))), max(abs(py_token_pe.mean(0))))     # max(max(px_token_pe.mean(0)), max(py_token_pe.mean(0)))
    ymin = -ymax  # min(min(px_token_pe.mean(0)), min(py_token_pe.mean(0)))
    
    return ymin*1.1, ymax*1.1
    

def plot_all_token_pe(
    dative_df, 
    figsize=None, 
    axes=None, 
    fig=None,
    model_name=None, 
    condition: str = '', 
    ylim=None, 
    save_to=None, 
    skip_first_n=0,
    tight_title=False,
    plot_ylabel=True,
    po_colors=None,
    do_colors=None,
    yticks=None,
    plot_delta_region=False,
    suptitle=None,
):
    if axes is None:
        fig, axes = plt.subplots(1, 2, figsize=figsize)

    po_title = "PO ($s$-PE: %pe)" if tight_title else "$w$-PE(PO) ($s$-PE: %pe)"
    do_title = "DO ($s$-PE: %pe)" if tight_title else "$w$-PE(DO) ($s$-PE: %pe)"
    
    example_sen_po = None if tight_title else "The girl gave the ball to the boy ."
    example_sen_do = None if tight_title else "The girl gave the boy the ball ."

    avg_df_po, subset_df_po = create_token_pe_plot(
        axes[0],
        dative_df,
        "all_logp_x_px",
        "all_logp_x_py",
        ["DT$_1$", "N$_1$", "V$_1$", "DT$_2$", "N$_2$", "P$_1$", "DT$_3$", "N$_3$", "."],
        po_title,
        ylim=ylim,
        skip_first_n=skip_first_n,
        plot_ylabel=plot_ylabel,
        tight_title=tight_title,
        colors=po_colors,
        yticks=yticks,
        plot_delta_region=plot_delta_region,
        label="A.",
        example_sen=example_sen_po,
    )

    avg_df_do, subset_df_do = create_token_pe_plot(
        axes[1],
        dative_df,
        "all_logp_y_py",
        "all_logp_y_px",
        ["DT$_1$", "N$_1$", "V$_1$", "DT$_2$", "N$_2$", "DT$_3$", "N$_3$", "."],
        do_title,
        num_pad=1,
        ylim=ylim,
        plot_ylabel=False,
        skip_first_n=skip_first_n,
        tight_title=tight_title,
        colors=do_colors,
        yticks=yticks,
        plot_delta_region=plot_delta_region,
        label="B.",
        example_sen=example_sen_do,
    )

    plt.subplots_adjust(wspace=.07, hspace=.5)
    
    if fig is not None and suptitle is not None:
        fig.suptitle(suptitle, y=1.15, fontsize=14)
    
    if save_to is not None:
        plt.savefig(save_to, bbox_inches='tight')
        plt.show()
    
    return avg_df_po, avg_df_do, subset_df_po, subset_df_do
```

```python
def compute_lmm_r2(lmem, result, true_result):
    # https://stats.stackexchange.com/questions/467543/including-random-effects-in-prediction-with-linear-mixed-model
    re = result.random_effects

    # Multiply each BLUP by the random effects design matrix for one group
    rex = [np.dot(lmem.exog_re_li[j], re[k]) for (j, k) in enumerate(lmem.group_labels)]

    # Add the fixed and random terms to get the overall prediction
    rex = np.concatenate(rex)
    yp = result.fittedvalues + rex

    return r2_score(true_result, yp)
```
